```python
#丟骰子遊戲,4個骰子
#2個要相同,才算另外2個分數

import pyinputplus as pyip
import random
import array as arr

dice=[0,0,0,0]#初始骰子都設0點
y=[0,0,0,0,0,0,0]#7個 zero array
point=0
def sido():
    dice[0]=random.randint(1,6)
    dice[1]=random.randint(1,6)
    dice[2]=random.randint(1,6)
    dice[3]=random.randint(1,6)
    print("4 dice = ",dice)
    return dice

def sido_ok():
    ok_sign=False
    global y
    y=[0,0,0,0,0,0,0]
    point=0
    x = arr.array('i', dice)
    print(x)
    for i in range(1,6):
        y[i]=x.count(i)
        if y[i]>=2:
            point=i
            print("第",i,"數有2個以上")
            #SHOW point positive
            ok_sign=True
        print(y)    
    return ok_sign

def print_sido_number():
    print(dice)
    sum=0
    for i in range(1,6):
        if y[i]==2 :
            for j in range(i+1,6):
                if y[j]==1:
                    sum+=j
                elif y[j]==3:
                    sum+=j
                elif y[j]==4:
                    sum=sum+j+j
                else:#2個dice 是J
                    continue
            for j in range(1,i-1):
                if y[j]==1:
                    sum+=j
                elif y[j]==3:
                    sum+=j
                elif y[j]==4:
                    sum=sum+j+j
                else:#2個dice 是J
                    continue
    print ("SUM =", sum)




print("============置股子遊戲==============")
while True:
    go=pyip.inputStr("請置骰子 : yes/no(y/n)")
    print(go)
    if go=="y":
        while True:
            dice1=sido()
            if sido_ok():
                print(dice)
                print_sido_number()
                break
            else:
                continue
 
    else:
        print("恭喜發財,BYE")
        break
```
